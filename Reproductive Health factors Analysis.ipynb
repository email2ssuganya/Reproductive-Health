```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url = "https://raw.githubusercontent.com/email2ssuganya/Reproductive-Health/main/Public_Health_Birth_Data__Pregnancy_Health_and_Risk_Factors.csv"

df = pd.read_csv(url)
df.head()
```

```python
print(df.shape)
print(df.columns)
print(df.info())

# Missing values
print(df.isnull().sum())
```

This gives insight about the columns available in the dataset

```python
df.replace("*", np.nan, inplace=True)

# Drop missing values (simple approach)
df.dropna(inplace=True)

# Reset index
df.reset_index(drop=True, inplace=True)
```

```python
# Missing values
print(df.isnull().sum())
```

This is the data lookup after removing duplicates

```python
numeric_df = df.select_dtypes(include=[np.number])
print(numeric_df.shape)
print(numeric_df.head())
print(df.columns)
for col in df.columns:
    if 'Count' in col or 'Rate' in col:
        df[col] = pd.to_numeric(df[col], errors='coerce')
numeric_df = df.select_dtypes(include=[np.number])
print(numeric_df.shape)
print(numeric_df.head())
```

```python
numeric_df = numeric_df.drop('Year', axis=1, errors='ignore')
top_features = numeric_df.sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(10,5))
top_features.plot(kind='bar')
plt.title("Top 10 Risk Factors")
plt.xticks(rotation=45)
plt.show()
```

Chart Description: Top 10 Risk Factors

This bar chart represents the top 10 risk factors based on their total counts/rates.

🔹 Key Insights:
Low Weight Gain is the highest risk factor

Followed by:

Inadequate Prenatal Care

Other Risks

Moderate contributors:

Hypertension

Diabetes

Lower contributors:

Smoking

Rate-based factors (very small values)

```python
plt.figure(figsize=(12,8))
sns.heatmap(numeric_df.corr(), cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
```

***🔍 Key Insights:***
1. Hypertension & Smoking
Mostly reddish tones
👉 Suggests a positive relationship
➡️ People who smoke more may have higher hypertension rates
2. Count vs Rate
Some variation in colors
👉 Means count and rate don’t always move together strongly
3. Blue Area (Negative Correlation)
That blue patch indicates:
👉 One variable increases while the other decreases

***📊 What it means:***

“This heatmap shows the correlation between health-related variables like hypertension and smoking. Red indicates strong positive correlation, blue indicates negative correlation. For example, smoking appears positively correlated with hypertension, suggesting a possible relationship between lifestyle and health conditions.”

***Summary:***

👉 Correlation ≠ Causation
Just because smoking and hypertension are correlated doesn’t mean one directly causes the other (though medically it might, but data alone doesn’t prove it).

```python
df.filter(like='Count').sum().sort_values(ascending=False).plot(kind='bar')
plt.title("Top Pregnancy Risk Factors")
plt.show()
```

***🔍 Key Insights:***

Excess Weight Gain is the highest risk factor (around 500,000 cases)
Lack of Prenatal Care is the second biggest issue
Other Risks (combined factors) also contribute significantly
Hypertension and Diabetes are moderate risks
Smoking and Alcohol Consumption are least reported, but still important

***Summary :***

Weight gain and lack of prenatal care are the most common pregnancy risk factors, while lifestyle habits like smoking and alcohol are less frequent but still critical.

```python
df.groupby('Year')['Count of Diabetes'].sum().plot()
plt.title("Diabetes Trend Over Years")
plt.show()
```

***🔍 Key Insights:***

From 2015 to 2016, diabetes cases show a slight increase
In 2018, there is a sharp drop (lowest point in the chart)
From 2019 to 2020, cases rise rapidly, reaching the peak in 2020
In 2021, there is a noticeable decline
From 2022 to 2023, the trend stabilizes with a slight decrease

***📊 Overall Trend:***

The data shows fluctuations, not a steady increase or decrease
2020 is the highest point, while 2018 is the lowest

***Summary:***

Diabetes cases fluctuate over the years, peaking in 2020 after a sharp drop in 201

```python
df.filter(like='Rate').mean().sort_values(ascending=False).plot(kind='bar')
plt.title("Average Risk Rates")
plt.show()
```

***🔍 Key Insights:***

Excess Weight Gain has the highest average risk rate (~3.3)

Lack of Prenatal Care (~2.1)

 and Other Risks (~1.9) follow

Diabetes (~0.9) and

 Hypertension (~0.9) are moderate risks

Smoking (~0.1) and

Alcohol Consumption (~0.02) are very low

📊 What it means:

Health and care-related factors (weight & prenatal care) contribute the most risk

Lifestyle habits (smoking, alcohol) show lower average impact in this dataset

Medical conditions (diabetes, hypertension) are important but mid-level

***Summary:***

Weight gain and prenatal care have the highest average risk rates, while smoking and alcohol have minimal impact in comparison.
